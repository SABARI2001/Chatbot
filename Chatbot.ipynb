```python
import re

# Define the math functions
def add(num1, num2):
    return num1 + num2

def subtract(num1, num2):
    return num1 - num2

def multiply(num1, num2):
    return num1 * num2

def divide(num1, num2):
    if num2 == 0:
        raise ZeroDivisionError("Cannot divide by zero!")
    else:
        return num1 / num2

# Define a dictionary of operations and their associated functions
operations = {
    "addition": add,
    "subtraction": subtract,
    "multiplication": multiply,
    "division": divide
}

# Function to get valid input from user
def get_valid_input(prompt):
    while True:
        user_input = input(prompt).strip()
        if user_input.lower() == 'exit':
            return None
        elif re.match("^[0-9]*\.?[0-9]+$", user_input):
            return float(user_input)
        else:
            print("Bot: Please enter a valid number!")
            
# Function to perform the math operation
def do_math_operation():
    print("Bot: Hi there! What mathematical operation do you want to perform?")
    while True:
        # Get the operation from the user
        while True:
            operation = input("User: ").strip()
            if operation.lower() == 'exit':
                return
            elif operation.lower() in operations:
                break
            else:
                print("Bot: Sorry, I don't know how to do that. Please choose one of the following operations:")
                print(list(operations.keys()))
        # Get the operands from the user
        try:
            num1 = get_valid_input("Bot: What is the first operand? ")
            if num1 is None:
                return
            num2 = get_valid_input("Bot: What is the second operand? ")
            if num2 is None:
                return
            # Perform the operation
            result = operations[operation.lower()](num1, num2)
            print(f"Bot: The result of {operation.lower()} of {num1} and {num2} is {result}")
        except ValueError:
            print("Bot: Please enter a valid number!")
        except ZeroDivisionError as e:
            print(f"Bot: {e}")
        print("Bot: Do you want to continue, start another operation, or exit?")
        while True:
            user_input = input("User: ").strip()
            if user_input.lower() in ['continue', 'start another operation', 'exit']:
                break
            else:
                print("Bot: I didn't understand. Please enter 'continue', 'start another operation', or 'exit'.")
        if user_input.lower() == 'exit':
            return
        elif user_input.lower() == 'start another operation':
            print("Bot: What mathematical operation do you want to perform?")

# Call the main function
do_math_operation()
```
